```python
def cria_plot():
    x = np.random.randn(5000, 6)
    (figure, axes) = plt.subplots(figsize = (16,10))
    (n, bins, patches) = axes.hist(x, 
                                    12, 
                                    normed = 1, 
                                    histtype = 'bar',
                                    label = ['Color 1', 'Color 2', 'Color 3', 'Color 4', 'Color 5']
                                  )
    axes.set_title("Histograma\npara\nDistribuicao Normal", fontsize = 25)
    axes.set_xlabel("Dados", fontsize = 16)
    axes.set_ylabel("Frequencia", fontsize = 16)
    axes.legend()
    plt.show()
```

```python
cria_plot()
```
