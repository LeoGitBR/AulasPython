```python
from scipy import stats
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Precos de casas
x = np.array([122,345,199,487,593,301,491,359])
y = np.array([1112, 1522, 2121, 2352, 2599, 3201, 3630, 4424])
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
```

```python
plt.plot(y, y, 'ro', color=  'red')
```

```python
p_value
```

```python
plt.ylabel('Preco')
plt.xlabel('Tamanho da casa')
plt.axis([0, 600, 0, 5000])
plt.plot(x, y);
plt.plot(x, x * slope + intercept, 'b')
```
